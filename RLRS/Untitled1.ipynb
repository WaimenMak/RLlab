```python
import numpy as np
x = np.array([[0.093, 0.1939, -1.0649, 0.4476, -2.0769],
[-1.8024, 0.3696, 0.7796, -1.0346, 0.473],
[0.5593, -2.5067, -2.1275, 0.5548, -1.6639]])

y = np.array([1, 2, 3])
```

```python
def softmax(x):
    t = np.exp(x)
    for i in range(len(x)):
        t[i,:] = t[i,:]/np.sum(np.exp(x[i,:]))
    return t

def cross_entropy(x, y):
	Y = np.zeros([len(y), x.shape[1]])
	for i in range(len(y)):
		Y[i, y[i]] = 1

	log_softmax = np.log(softmax(x))
	cross_ent = 0;
	cross_ent = np.sum(np.dot(log_softmax, Y.T))
	return -cross_ent
```
